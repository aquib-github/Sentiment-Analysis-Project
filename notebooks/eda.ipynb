# 🔍 Exploratory Data Analysis — Twitter Sentiment Dataset

This notebook explores the raw dataset before model training.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud

# Load dataset
df = pd.read_csv('../data/raw/twitter_training.csv', header=None,
                 names=['id', 'topic', 'sentiment', 'text'])
df.dropna(inplace=True)
df['text'] = df['text'].astype(str)
print(f'Shape: {df.shape}')
df.head()
```

```python
# Sentiment distribution
print(df['sentiment'].value_counts())
df['sentiment'].value_counts().plot(kind='bar', color=['#2ecc71', '#e74c3c', '#3498db', '#f39c12'])
plt.title('Sentiment Distribution')
plt.ylabel('Count')
plt.show()
```

```python
# Text length distribution
df['text_length'] = df['text'].apply(len)
df['text_length'].hist(bins=50, figsize=(10, 4), color='#3498db', edgecolor='black')
plt.title('Distribution of Tweet Lengths')
plt.xlabel('Character Count')
plt.ylabel('Frequency')
plt.show()
```

```python
# Word cloud for all text
all_text = ' '.join(df['text'].values)
wc = WordCloud(width=800, height=400, background_color='white', max_words=150).generate(all_text)
plt.figure(figsize=(12, 5))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud — All Tweets', fontsize=14, fontweight='bold')
plt.show()
```

```python
# Topic distribution
print(f'Unique topics: {df["topic"].nunique()}')
df['topic'].value_counts().head(20).plot(kind='barh', figsize=(10, 6), color='#9b59b6')
plt.title('Top 20 Topics')
plt.xlabel('Count')
plt.show()
```
